```python
import os
import shutil
import random

# --- 설정 (Configuration) ---
# 원본 우동 이미지 파일이 있는 디렉토리
# 이 경로를 당신의 실제 경로로 정확히 변경해주세요.
SOURCE_IMAGE_DIR = 'E:\★★★★★AI★★★★★\음식 ai data\selectStart음식DATA\Computer Vision Lab\ramen\ramen\png'

# 분할된 데이터를 저장할 상위 디렉토리 (자동 생성됩니다)
# 이 경로도 당신이 원하는 곳으로 변경해주세요.
# 예: SOURCE_IMAGE_DIR의 상위 폴더에 'udon_split_dataset'을 만드는 경우
DESTINATION_ROOT_DIR = 'E:/★★★★★AI★★★★★/음식 ai data/selectStart음식DATA/Computer Vision Lab/ramen_split_dataset'

# 각 세트의 비율 (합이 1.0이 되어야 합니다)
TRAIN_RATIO = 0.8
VALIDATION_RATIO = 0.1
TEST_RATIO = 0.1

# (선택 사항) 재현성을 위한 시드 설정
# 이 숫자를 고정하면, 코드를 여러 번 실행해도 항상 동일한 분할 결과를 얻습니다.
RANDOM_SEED = 42
random.seed(RANDOM_SEED)

# --- 폴더 생성 함수 ---
def create_split_directories(root_dir, class_name):
    """훈련, 검증, 테스트 폴더 구조를 생성합니다."""
    train_dir = os.path.join(root_dir, 'train', class_name)
    validation_dir = os.path.join(root_dir, 'validation', class_name)
    test_dir = os.path.join(root_dir, 'test', class_name)

    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(validation_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    print(f"Created directories:\n- {train_dir}\n- {validation_dir}\n- {test_dir}")
    return train_dir, validation_dir, test_dir

# --- 파일 분할 및 복사 함수 ---
def split_and_copy_files(source_dir, dest_root_dir, train_ratio, val_ratio, test_ratio):
    print(f"Reading files from: {source_dir}")
    # 원본 디렉토리의 모든 파일 목록 가져오기
    all_files = [f for f in os.listdir(source_dir) if os.path.isfile(os.path.join(source_dir, f))]

    # 이미지 파일 (확장자 .png만) 필터링
    image_files = [f for f in all_files if f.lower().endswith('.png')]

    if not image_files:
        print(f"Error: No .png image files found in {source_dir}. Please check the path and file extensions.")
        return

    # 파일 목록 섞기 (랜덤하게)
    random.shuffle(image_files)

    num_files = len(image_files)
    num_train = int(num_files * train_ratio)
    num_validation = int(num_files * val_ratio)
    # 나머지는 테스트 세트로 할당하여 모든 파일이 포함되도록 합니다.
    num_test = num_files - num_train - num_validation

    # 파일 목록 분할
    train_files = image_files[:num_train]
    validation_files = image_files[num_train : num_train + num_validation]
    test_files = image_files[num_train + num_validation :]

    print(f"\nTotal .png files found: {num_files}")
    print(f"Allocating {len(train_files)} files for training ({train_ratio*100:.0f}%)")
    print(f"Allocating {len(validation_files)} files for validation ({val_ratio*100:.0f}%)")
    print(f"Allocating {len(test_files)} files for testing ({test_ratio*100:.0f}%)")

    # 대상 폴더 구조 생성
    # 'udon' 폴더의 이름이 클래스 이름이 됩니다.
    class_name = os.path.basename(source_dir) # 'udon' 추출
    train_dest_dir, val_dest_dir, test_dest_dir = create_split_directories(dest_root_dir, class_name)

    # 파일 복사
    print("\nStarting file copying...")

    # 훈련 세트 복사
    for i, filename in enumerate(train_files):
        src_path = os.path.join(source_dir, filename)
        dest_path = os.path.join(train_dest_dir, filename)
        shutil.copy(src_path, dest_path)
        if (i + 1) % 100 == 0: # 100개마다 진행 상황 출력
            print(f"Copied {i + 1}/{len(train_files)} train files...")
    print(f"Finished copying {len(train_files)} train files to {train_dest_dir}")

    # 검증 세트 복사
    for i, filename in enumerate(validation_files):
        src_path = os.path.join(source_dir, filename)
        dest_path = os.path.join(val_dest_dir, filename)
        shutil.copy(src_path, dest_path)
        if (i + 1) % 100 == 0:
            print(f"Copied {i + 1}/{len(validation_files)} validation files...")
    print(f"Finished copying {len(validation_files)} validation files to {val_dest_dir}")

    # 테스트 세트 복사
    for i, filename in enumerate(test_files):
        src_path = os.path.join(source_dir, filename)
        dest_path = os.path.join(test_dest_dir, filename)
        shutil.copy(src_path, dest_path)
        if (i + 1) % 100 == 0:
            print(f"Copied {i + 1}/{len(test_files)} test files...")
    print(f"Finished copying {len(test_files)} test files to {test_dest_dir}")

    print("\nFile splitting and copying complete!")
    print(f"Check your new dataset structure at: {DESTINATION_ROOT_DIR}")

# --- 스크립트 실행 ---
if __name__ == "__main__":
    split_and_copy_files(SOURCE_IMAGE_DIR, DESTINATION_ROOT_DIR, TRAIN_RATIO, VALIDATION_RATIO, TEST_RATIO)
```

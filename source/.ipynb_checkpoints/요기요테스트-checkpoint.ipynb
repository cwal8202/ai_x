```python
import re
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# 크롬 드라이버 경로 설정
driver = webdriver.Chrome()
driver.get('https://www.yogiyo.co.kr/')
time.sleep(0.1)  # 페이지 로딩 대기 (필요 시 더 길게 조정)

elem = driver.find_element(By.CLASS_NAME, 'btn.btn-default.ico-loc').click()
time.sleep(0.1)
```

```python
# 따로 땔 줄 몰라서 일단 분리 실행
elem = driver.find_element(By.CLASS_NAME, 'item.clearfix').click()
time.sleep(0.1)
```

```python
# 렌더링된 페이지 소스 가져오기
soup = BeautifulSoup(driver.page_source, 'html.parser')

# 이미지 div 추출
photo_divs = soup.find_all('div', class_='photo')

for div in photo_divs:
    style = div.get('style', '')
    image_urls = re.findall(r"url\('([^']+)'\)", style)
    for idx, image_url in enumerate(image_urls):
        print(f"이미지 URL: {image_url}")
driver.quit()
```
